```python
import pandas as pd
```

```python
df = pd.read_csv("forestdata.csv")
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df
```

```python
df.isnull().sum()
```

```python
text_to_number = {
    'morning': 0,
    'afternoon': 1,
    'night': 2
}

# Replace text values with numerical values in a specific column (assuming column name is 'Time')
df['time.of.day'] = df['time.of.day'].replace(text_to_number)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(columns=['fire'])
y = df['fire']
```

```python
X
```

```python
y
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y, test_size=0.5, random_state=42)
```

# **logistic regression model**

```python
# prompt: train lr model

from sklearn.linear_model import LogisticRegression

# Create a Logistic Regression model
model = LogisticRegression()

# Train the model using the training data
model.fit(X_train, y_train)
```

```python
# prompt: print acc
print(" test accuracy :")
print("Accuracy:", model.score(X_test, y_test))
print(" Train accuracy :")
print("Accuracy:", model.score(X_train, y_train))
```

# **Random forest**

```python
X_train,X_test,y_train,y_test=train_test_split(X,y, test_size=0.3, random_state=42)
```

```python
# prompt: train random forest model
from sklearn.ensemble import RandomForestClassifier

# Define hyperparameters
n_estimators = 3
max_depth = None
min_samples_split = 2
min_samples_leaf = 12

# Create a Random Forest Classifier model with specified hyperparameters
model = RandomForestClassifier(n_estimators=n_estimators,
                               max_depth=max_depth,
                               min_samples_split=min_samples_split,
                               min_samples_leaf=min_samples_leaf)

# Train the model using the training data
model.fit(X_train, y_train)


print(" test accuracy :")
print("Accuracy:", model.score(X_test, y_test))
print(" Train accuracy :")
print("Accuracy:", model.score(X_train, y_train))
```

# **decision trees**

```python


from sklearn.tree import DecisionTreeClassifier


max_depth = 5
min_samples_split = 4
min_samples_leaf = 10


model = DecisionTreeClassifier(max_depth=max_depth,
                               min_samples_split=min_samples_split,
                               min_samples_leaf=min_samples_leaf)


model.fit(X_train, y_train)


print(" test accuracy :")
print("Accuracy:", model.score(X_test, y_test))
print(" Train accuracy :")
print("Accuracy:", model.score(X_train, y_train))
```

# KNN **model**

```python


from sklearn.neighbors import KNeighborsClassifier
import joblib


n_neighbors = 2


model = KNeighborsClassifier(n_neighbors=n_neighbors)


model.fit(X_train, y_train)

print(" test accuracy :")
print("Accuracy:", model.score(X_test, y_test))
print(" Train accuracy :")
print("Accuracy:", model.score(X_train, y_train))
import joblib

# Assuming you have already trained your KNN model and named it 'model'

# Save the model to a file
joblib.dump(model, 'knn_model.pkl')
```

```python
import numpy as np
import joblib


# Load the saved model
model = joblib.load('/content/knn_model.pkl')

# Function to get user input for each attribute
def get_input(attribute_name):
    while True:
        try:
            value = float(input("Enter {} value: ".format(attribute_name)))
            return value
        except ValueError:
            print("Invalid input! Please enter a numerical value.")

# Get user input for each attribute
collector_id = get_input("collector_id")
c_score = get_input("c_score")
l_score = get_input("l_score")
rain = get_input("rain")
tree_age = get_input("tree_age")
surface_litter = get_input("surface_litter")
wind_intensity = get_input("wind_intensity")
humidity = get_input("humidity")
tree_density = get_input("tree_density")
month = get_input("month")
time_of_day = get_input("time_of_day")

# Create new data array for prediction
new_data = np.array([[collector_id, c_score, l_score, rain, tree_age, surface_litter, wind_intensity, humidity, tree_density, month, time_of_day]])

# Make predictions on the new data
prediction = model.predict(new_data)

# Print the prediction
print("Predicted class:", prediction[0])
```
